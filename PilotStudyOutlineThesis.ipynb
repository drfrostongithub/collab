```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = '/content/drive/MyDrive/Pilot_Study.csv'  # Replace with your file's actual path
df = pd.read_csv(file_path)
ai_handling = df['When you receive AI-generated code output, how do you typically handle it?'].value_counts()
productivity_impact = df['If AI were removed from your workflow, how would it affect your productivity?'].value_counts()

fig, axes = plt.subplots(1, 2, figsize=(12, 6))  # 1 row, 2 columns of subplots

# Bar chart for AI handling with percentages
# Calculate percentages for AI handling
ai_percentages = (ai_handling / ai_handling.sum() * 100).round(2)

bars = axes[0].bar(ai_handling.index, ai_handling.values) # Get the bar objects
axes[0].set_title('Handling of AI-Generated Code')
axes[0].set_xlabel('Response')
axes[0].set_ylabel('Percentage')

# Add percentage labels above each bar
for bar, percentage in zip(bars, ai_percentages):
    height = bar.get_height()
    axes[0].text(bar.get_x() + bar.get_width() / 2, height + 1, f'{percentage:.2f}%',
                 ha='center', va='bottom')

# Rotate x-axis labels and align them to the right
axes[0].tick_params(axis='x', labelrotation=45)
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45, ha='right')


# Bar chart for productivity impact with percentages
# Calculate percentages for productivity impact
productivity_percentages = (productivity_impact / productivity_impact.sum() * 100).round(2)

bars = axes[1].bar(productivity_impact.index, productivity_impact.values) # Get the bar objects
axes[1].set_title('Impact of AI Removal on Productivity')
axes[1].set_xlabel('Response')
axes[1].set_ylabel('Percentage')

# Add percentage labels above each bar
for bar, percentage in zip(bars, productivity_percentages):
    height = bar.get_height()
    axes[1].text(bar.get_x() + bar.get_width() / 2, height + 1, f'{percentage:.2f}%',
                 ha='center', va='bottom')

# Rotate x-axis labels and align them to the right
axes[1].tick_params(axis='x', labelrotation=45)
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()  # Adjust spacing
plt.show()
```

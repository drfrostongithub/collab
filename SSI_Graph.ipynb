```python
from graphviz import Digraph
from IPython.display import Image
from google.colab import files
import os

# Membuat objek flowchart
flowchart = Digraph("Boeing_Production_Process", format="png")

# Menambahkan node
flowchart.node("A", "Perencanaan Produksi\n(Menentukan jumlah pesawat)")
flowchart.node("B", "Pembelian Bahan Baku\n(Aluminium, komposit, mesin)")
flowchart.node("C", "Pabrikasi Komponen\n(Sayap, badan pesawat, avionik)")
flowchart.node("D", "Perakitan Pesawat\n(Menggabungkan semua komponen)")
flowchart.node("E", "Pengujian Kualitas\n(Uji coba keselamatan dan kinerja)")
flowchart.node("F", "Pengiriman ke Pelanggan\n(Diserahkan ke maskapai)")

# Menambahkan alur proses
flowchart.edge("A", "B")
flowchart.edge("B", "C")
flowchart.edge("C", "D")
flowchart.edge("D", "E")
flowchart.edge("E", "F")

# Menyimpan dan menampilkan flowchart
# flowchart_path = "/mnt/data/Boeing_Production_Process.png"
flowchart_path = "Boeing_Production_Process.jpg"
flowchart.render(flowchart_path, format="jpg", cleanup=False)
flowchart_path

# Display the flowchart within the notebook
display(Image(filename=flowchart_path))

# Download the flowchart
files.download(flowchart_path)
```

```python
import matplotlib.pyplot as plt

# Data
aircraft_produced = 400  # Number of Boeing 737 MAX produced per year
transactions = 400      # Number of delivery transactions

# Create a bar chart
plt.bar(["Production", "Transactions"], [aircraft_produced, transactions])

# Add labels and title
plt.xlabel("Process")
plt.ylabel("Quantity")
plt.title("Boeing 737 MAX Production and Transaction Volume (Annual)")

# Display the chart
plt.show()
```

```python
from graphviz import Digraph

# Create a new Digraph object
infra_flowchart = Digraph("IT_Infrastructure", format="png")

# Add nodes for each component
infra_flowchart.node("Data Center", "Data Center\n(Pusat Data)")
infra_flowchart.node("Cloud", "Cloud Computing\n(AWS, Azure)")
infra_flowchart.node("ERP", "ERP System\n(SAP)")
infra_flowchart.node("CAD/CAM", "CAD/CAM Software\n(Desain Pesawat)")
infra_flowchart.node("IoT", "IoT\n(Monitoring Produksi)")
infra_flowchart.node("Cybersecurity", "Cybersecurity System\n(Keamanan Data)")

# Add edges to represent the connections
infra_flowchart.edge("Data Center", "Cloud")
infra_flowchart.edge("Cloud", "ERP")
infra_flowchart.edge("ERP", "CAD/CAM")
infra_flowchart.edge("CAD/CAM", "IoT")
infra_flowchart.edge("IoT", "Cybersecurity")

# Render and save the flowchart
infra_flowchart.render("IT_Infrastructure", view=True)
```

```python
from graphviz import Digraph

# Create a new Digraph object
future_flowchart = Digraph("Future_Production_Process", format="png")

# Add nodes for each step with technology integration
future_flowchart.node("A", "Perencanaan Produksi\n(Dengan AI)")
future_flowchart.node("B", "Pembelian Bahan Baku\n(Otomatis via Blockchain)")
future_flowchart.node("C", "Pabrikasi Komponen\n(Robotik dan 3D Printing)")
future_flowchart.node("D", "Perakitan Pesawat\n(Augmented Reality)")
future_flowchart.node("E", "Pengujian Kualitas\n(Drone dan Sensor IoT)")
future_flowchart.node("F", "Pengiriman ke Pelanggan\n(Drone dan Autonomous Vehicles)")

# Add edges to represent the flow
future_flowchart.edge("A", "B")
future_flowchart.edge("B", "C")
future_flowchart.edge("C", "D")
future_flowchart.edge("D", "E")
future_flowchart.edge("E", "F")

# Render and save the flowchart
future_flowchart.render("Future_Production_Process", view=True)
```
